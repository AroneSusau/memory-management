```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
%config InlineBackend.figure_format = 'retina'
plt.rcParams['figure.figsize'] = (10, 6)
plt.style.use('ggplot')
pd.set_option('precision', 2)
sns.set_theme(color_codes=True)

```

```python
data = pd.read_csv('./combined.csv')
```

```python
list_type = (data["list"] == "FREE")

free = data[list_type]

sns.catplot(x="allocator", y="size", hue="allocator", kind="box", dodge=False, data=free)
```

```python
totals = pd.read_csv('./totals.csv')

sns.barplot(data=totals, y="allocator", x="memory_usage", hue="file")
```

```python
sns.boxplot(x="allocator", y="size", palette=["m", "g"], data=free)
sns.despine(offset=10, trim=True)
```
